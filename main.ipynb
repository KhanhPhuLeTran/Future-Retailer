# Pre-processing

### Ghép file

```python
sale_df = pd.concat([dataframes['10.SE Regional Sales'], dataframes['11.SE Regional Sales'], dataframes['12.SE Regional Sales']], axis=0)
sale_df
```

### Chuyển đổi tên các bang

```python
dataframes['Customer']['State'].value_counts()
```

```python
mapper = {
    'LA': 'Louisiana',
    'AL': 'Alabama',
    'SC': 'South Carolina',
   'NC': 'North Carolina',
    'FL': 'Florida',
    'GA': 'Georgia'
}

dataframes['Customer']['State'] = dataframes['Customer']['State'].map(mapper)
dataframes['Customer']['State'].value_counts()
```

### Tạo cột `customer_name` từ 2 cột `first_name` và `last_name`

```python
dataframes['Customer']['Customer Name'] = dataframes['Customer']['Customer First Name'] + ' ' + dataframes['Customer']['Customer Last Name']
# remove customer first name and last name
dataframes['Customer'] = dataframes['Customer'].drop(['Customer First Name', 'Customer Last Name'], axis=1)
dataframes['Customer'].head()
```

```python
# Move the 'Customer_name' column to the third column
cols = list(dataframes['Customer'].columns)
cols = cols[:2] + [cols[-1]] + cols[2:-1]
dataframes['Customer'] = dataframes['Customer'][cols]
dataframes['Customer'].head()
```

### Chuyển đổi kiểu dữ liệu 

```python
sale_df['Order ID'] = sale_df['Order ID'].astype(str)
```

```python
sale_df['Customer ID'] = sale_df['Customer ID'].astype(str)
```

### Tạo thêm 3 cột mới từ cột `Order Date`

```python
sale_df['Order Date']
```

### Bỏ cột `State` khỏi bảng `Regional Sales`

```python
sale_df
```

```python
sale_df = sale_df.drop(['State', 'Customer Last Name', 'Customer First Name', 'Total Profit '], axis=1)
```

```python
dataframes['Customer'] = dataframes['Customer'].drop(['Product Category', 'Order Date', 
                                                      'Quantity Ordered', 'Order Status', 'Sale Price'], axis=1)
```

### Tạo thêm 1 bảng mới `Postal Code Mapping`

```python
# lấy 3 cột từ dataframes['Customer'] thành 1 dataframe mới
postal_df = dataframes['Customer'][['Postal Code', 'City', 'State']]
dataframes['Customer'] = dataframes['Customer'].drop(['City', 'State'], axis=1)
```

```python
postal_df = postal_df.drop_duplicates()
```

```python
dataframes['Customer'] = dataframes['Customer'].drop_duplicates(subset=['Customer ID'], keep='first')
```

### Lưu các bảng đã xử lý

```python
postal_df = postal_df.drop_duplicates(subset=['Postal Code'], keep='first')
```

```python
def plot_correlation_heatmap_with_categorical(dataframe, figsize=(10, 8)):
    """
    Hiển thị biểu đồ heatmap của ma trận tương quan giữa các cột, bao gồm cả cột phân loại trong DataFrame.

    Parameters:
    dataframe (pd.DataFrame): DataFrame chứa dữ liệu
    figsize (tuple): Kích thước của biểu đồ (mặc định là (10, 8))

    Returns:
    None
    """
    # Sao chép DataFrame để tránh thay đổi dữ liệu gốc
    df_encoded = sale_df.copy()

    # Mã hóa các cột phân loại bằng Label Encoding
    label_encoders = {}
    for column in df_encoded.select_dtypes(include=['object', 'category']).columns:
        le = LabelEncoder()
        df_encoded[column] = le.fit_transform(df_encoded[column])
        label_encoders[column] = le

    # Tính ma trận tương quan
    correlation_matrix = df_encoded.corr()

    # Thiết lập kích thước của biểu đồ
    plt.figure(figsize=figsize)

    # Vẽ heatmap
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5, linecolor='white')

    # Tiêu đề biểu đồ
    plt.title('Heatmap of Correlation Matrix with Categorical Data')

    # Hiển thị biểu đồ
    plt.show()

plot_correlation_heatmap_with_categorical(sale_df)
```
